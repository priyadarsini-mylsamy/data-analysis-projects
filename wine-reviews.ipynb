```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/home/pri/Downloads/winemag-data_first150k.csv",index_col=0)
fig_size = plt.rcParams["figure.figsize"]
```

Let us create a data frame for wine from US

```python
df_us = df.loc[df["country"] == 'US']
```

```python
df_us
```

Let's find the most expensive wine in US

```python
expensive_US_wine = df_us[["description","price"]].loc[df_us["price"]==max(df_us["price"])]
```

```python
expensive_US_wine
```

Let us find the cheapest wine in the US

```python
cheapest_US_wine=df_us[["description", "price","points"]].loc[df_us["price"]==min(df_us["price"])]
```

```python
cheapest_US_wine
```

List of the most expensive wines by country

```python
most_expensive_wine_countrywise=df[["country","price"]].groupby("country").max()
```

```python
most_expensive_wine_countrywise.dropna()
most_expensive_wine_countrywise.sort_values("price",ascending=False,inplace=True)
```

```python
most_expensive_wine_countrywise.reset_index(inplace=True)
```

```python
most_expensive_wine_countrywise_top15 = most_expensive_wine_countrywise.head(15)
```

```python
plt.bar(most_expensive_wine_countrywise_top15['country'],most_expensive_wine_countrywise_top15['price'])
plt.title('Most expensive wine by country')
plt.xlabel('Country')
plt.ylabel('Price')
fig_size[0]= 25
fig_size[1]=15
plt.rcParams["figure.figsize"] = fig_size
plt.show()
```

Distribution of number of wine between $0 to $100


```python
wine_price_lessthan100=df[["price"]].loc[df["price"] <= 100]
```

```python
wine_price_lessthan100.reset_index(inplace=True)
```

```python
plt.hist(wine_price_lessthan100['price'].values,bins=20)
plt.title("Distribution of Wine between 0 and $100")
plt.xlabel("Price")
plt.ylabel("Number of Wine")
plt.show()
```

```python
good_quality_wine=df[["winery","points"]].loc[df["points"]>=95 ]
```

```python
good_winery = good_quality_wine.groupby("winery").count()
print(good_winery)
```

```python
good_winery.reset_index(inplace=True)
```

```python
good_winery.sort_values("points",ascending=False,inplace=True)
```

```python
top_15_good_winery = good_winery.head(15)
```

```python
top_15_good_winery
```

```python
plt.bar(top_15_good_winery["winery"], top_15_good_winery["points"])
plt.title("Good Wineries")
plt.xlabel("winery")
plt.ylabel("count of good rated wine")

plt.show()
```

Points vs Price Corelation

```python
plt.scatter(df["price"],df["points"])
plt.title("Points-Price Corelation")
plt.xlabel("Price")
plt.ylabel("Points")
plt.show()
```

Let us look at 10 best wine producing provinces

```python
best_quality_wine=df[["province","points"]].loc[df["points"] > 98 ]
```

```python
best_quality_wine_byProvince=best_quality_wine.groupby("province").count()

```

```python
best_quality_wine_byProvince.reset_index(inplace=True)
```

```python
best_quality_wine_byProvince.sort_values("points",ascending=False,inplace=True)
```

```python
plt.bar(best_quality_wine_byProvince["province"].head(15),best_quality_wine_byProvince["points"].head(15))
plt.title("Best Wine Producing Province")
plt.xlabel("Province")
plt.ylabel("Number of best quality wines")
plt.show()
```
