# Wine review data analysis using pandas

**The Jupyter notebook sets out to explore a wine review data set to answer interesting questions (see below). The data set itself was downloaded from Kaggle and can be accessed [here](https://www.kaggle.com/zynicide/wine-reviews).** 

**Here, we will be using pandas to preview, analyze, manipulate and prepare the data set. For visualizations, we will be using the matplotlib library.**

To begin, below are some interesting questions that we can try to answer using the data set: 
1. Which is the most expensive wine within the US?
2. Which is the cheapest wine within the US?
3. Which are the most expensive wines, grouped by country? 
4. Can we create a distribution of wines between \$1 and \$100?
5. What are the top 15 wineries (by count of good rated wines)?
6. Is there a correlation between price and review points? 
7. What are the 10 best wine producing provinces (based on review points)? 
8. Which variety of grapes produces most of the best quality wines?


```python
# library imports
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# importing the data set from a csv file into a pandas dataframe named df_wine_reviewdata using read_csv method
df_wine_reviewdata = pd.read_csv("/home/pri/Downloads/winemag-data_first150k.csv",index_col=0)
fig_size = plt.rcParams["figure.figsize"]
```

We will now create a data frame for wine from the US only

```python
# filtering data for US
df_wine_reviewdata_us = df_wine_reviewdata.loc[df_wine_reviewdata["country"] == 'US']
```

Let's find the most expensive wine in US

```python
# selecting only price and description for the most expensive wine from the data frame
expensive_US_wine = df_wine_reviewdata_us[["description","price"]].loc[df_wine_reviewdata_us["price"]==
                                                                       max(df_wine_reviewdata_us["price"])]
```

```python
expensive_US_wine
```

Let us find the cheapest wine in the US

```python
# selecting only price and description for the cheapest wine from the data frame
cheapest_US_wine=df_wine_reviewdata_us[["description", "price","points"]].loc[df_wine_reviewdata_us["price"]==
                                                                              min(df_wine_reviewdata_us["price"])]
```

```python
# there are multiple results for the same price value
cheapest_US_wine
```

List of the most expensive wines by country

```python
# selecting country and price columns for the most expensive wine by country
most_expensive_wine_countrywise=df_wine_reviewdata[["country","price"]].groupby("country").max()
```

```python
# dropping null values and sorting by price
most_expensive_wine_countrywise.dropna()
most_expensive_wine_countrywise.sort_values("price",ascending=False,inplace=True)
```

```python
# resetting index to be used as a data frame column
most_expensive_wine_countrywise.reset_index(inplace=True)
```

```python
# selecting the top 15 values of the data frame
most_expensive_wine_countrywise_top15 = most_expensive_wine_countrywise.head(15)
```

```python
# visualization using matplotlib
plt.bar(most_expensive_wine_countrywise_top15['country'],most_expensive_wine_countrywise_top15['price'])
plt.title('Most expensive wine by country')
plt.xlabel('Country')
plt.ylabel('Price')
fig_size[0]= 30
fig_size[1]=20
plt.rcParams["figure.figsize"] = fig_size
plt.show()
```

Distribution of number of wine between $1 to $100


```python
# selecting wines that are priced less than $100
wine_price_lessthan100=df_wine_reviewdata[["price"]].loc[df_wine_reviewdata["price"] <= 100]
```

```python
# resetting index
wine_price_lessthan100.reset_index(inplace=True)
```

```python
# histogram visualization using matplotlib
plt.hist(wine_price_lessthan100['price'].values,bins=20)
plt.title("Distribution of Wine between 1 and $100")
plt.xlabel("Price")
plt.ylabel("Number of Wine")
plt.show()
```

```python
# wines at or over 95 points
good_quality_wine=df_wine_reviewdata[["winery","points"]].loc[df_wine_reviewdata["points"]>=95 ]
```

```python
# count of wines by winery
good_winery = good_quality_wine.groupby("winery").count()
```

```python
good_winery.reset_index(inplace=True)
```

```python
# sorting by points
good_winery.sort_values("points",ascending=False,inplace=True)
```

```python
# selecting the top 15 values
top_15_good_winery = good_winery.head(15)
```

```python
top_15_good_winery
```

```python
# visualization using matplotlib
plt.bar(top_15_good_winery["winery"], top_15_good_winery["points"])
plt.title("Good Wineries")
plt.xlabel("winery")
plt.ylabel("count of good rated wine")

plt.show()
```

Points vs Price Corelation

```python
# scatter plot visualization using matplotlib
plt.scatter(df_wine_reviewdata["price"],df_wine_reviewdata["points"])
plt.title("Points-Price Corelation")
plt.xlabel("Price")
plt.ylabel("Points")
plt.show()
```

Let us look at 10 best wine producing provinces

```python
# selecting provinces and points from the data frame
best_quality_wine=df_wine_reviewdata[["province","points"]].loc[df_wine_reviewdata["points"] > 98 ]
```

```python
# getting count by province
best_quality_wine_byProvince=best_quality_wine.groupby("province").count()
```

```python
best_quality_wine_byProvince.reset_index(inplace=True)
```

```python
# sort by points
best_quality_wine_byProvince.sort_values("points",ascending=False,inplace=True)
```

```python
# visualization using matplotlib
plt.bar(best_quality_wine_byProvince["province"].head(15),best_quality_wine_byProvince["points"].head(15))
plt.title("Best Wine Producing Province")
plt.xlabel("Province")
plt.ylabel("Number of best quality wines")
plt.show()
```
